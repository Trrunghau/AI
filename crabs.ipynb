```python
from pandas.compat import platform
import pandas as pd
url = 'crabs.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axis3d
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'width',y = 'index',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression() # hồi quy tuyến tính
model.fit(X=pd.DataFrame(data['width']),y=data['index']) # fit là benkyo
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['width']),y=data['index'])
data.plot(kind='scatter',x='width',y='index',figsize=(9,9),color='black',xlim=(0,55))
y_pred = model.predict(X=pd.DataFrame(data['width']))
plt.plot(data['width'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['width'],data['width']**2]).T
model.fit(X,y=data['index'])
print('Intercept của mô hình: ',model.intercept_)
print('Coef của mô hình:', model.coef_)
print('độ chính xác của mô hình:',model.score(X,y=data['index']))
import numpy as np

range = np.arange(0,55,0.3)
data.plot(kind='scatter',x='width',y='index',figsize=(9,9),color='black',xlim=(0,55))
pre=pd.DataFrame([range,range**2]).T    # bậc 2
y_pred=model.predict(X=pre)
plt.plot(range,y_pred,color='blue')
```
