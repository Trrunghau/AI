```python
from pandas.compat import platform
import pandas as pd
url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axis3d
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
data.plot(kind = 'scatter', x = 'Alkhol',y = 'Result',figsize=(9,9),color='black')
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression() # hồi quy tuyến tính
model.fit(X=pd.DataFrame(data['Alkhol']),y=data['Result']) # fit là benkyo
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['Alkhol']),y=data['Result'])
data.plot(kind='scatter',x='Alkhol',y='Result',figsize=(9,9),color='black',xlim=(0,10))
y_pred = model.predict(X=pd.DataFrame(data['Alkhol']))
plt.plot(data['Alkhol'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Alkhol'],data['Alkhol']**2]).T
model.fit(X,y=data['Result'])
print('Intercept của mô hình: ',model.intercept_)
print('Coef của mô hình:', model.coef_)
print('độ chính xác của mô hình:',model.score(X,y=data['Result']))
import numpy as np

range = np.arange(0,10,0.1)
data.plot(kind='scatter',x='Alkhol',y='Result',figsize=(9,9),color='black',xlim=(0,10))
pre=pd.DataFrame([range,range**2]).T    # bậc 2
y_pred=model.predict(X=pre)
plt.plot(range,y_pred,color='blue')
```
