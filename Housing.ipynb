```python
from pandas.compat import platform
import pandas as pd
url = 'Housing.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axis3d
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression() # hi quy tuyn tnh
model.fit(X=pd.DataFrame(data['area']),y=data['price']) # fit l benkyo
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['area']),y=data['price'])
data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='black',xlim=(0,15000))
y_pred = model.predict(X=pd.DataFrame(data['area']))
plt.plot(data['area'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['area'],data['area']**2]).T
model.fit(X,y=data['price'])
print('Intercept ca m hnh: ',model.intercept_)
print('Coef ca m hnh:', model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y=data['price']))
import numpy as np

range = np.arange(0,15000,0.5)
data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='black',xlim=(0,15000))
pre=pd.DataFrame([range,range**2]).T    # bc 2
y_pred=model.predict(X=pre)
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['area'],data['area']**2,data['area']**3]).T    # bc 3
model.fit(X,y=data['price'])
print('Intercept ca m hnh: ',model.intercept_)
print('Coef ca m hnh:', model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y=data['price']))

range = np.arange(0,15000,0.5)
data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='green',xlim=(0,15000))
pre=pd.DataFrame([range,range**2,range**3]).T    # bc 3
y_pred=model.predict(X=pre)
plt.plot(range,y_pred,color='blue')
```
