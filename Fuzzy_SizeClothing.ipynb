```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as crtl 

HeightA = crtl.Antecedent(np.arange(95,160,1),'HeightA')        # input
WeightA = crtl.Antecedent(np.arange(15,50,0.5),'WeightA') 
WaistA = crtl.Antecedent(np.arange(45,70,0.5),'WaistA')        # input
HeightB = crtl.Antecedent(np.arange(90,170,1),'HeightB') 
ChestB = crtl.Antecedent(np.arange(47,85,0.5),'ChestB')        # input
WaistB = crtl.Antecedent(np.arange(45,70,0.5),'WaistB') 

conclusionA = crtl.Consequent(np.arange(4,14,1),'conclusionA')    # output
conclusionB = crtl.Consequent(np.arange(4,14,0.5),'conclusionB') 

HeightA['5'] = fuzz.trimf(HeightA.universe,[98.6,111.94,116])
HeightA['6'] = fuzz.trimf(HeightA.universe,[98.6,118.7,123])
HeightA['6X'] = fuzz.trimf(HeightA.universe,[98.6,123.52,128])
HeightA['7'] = fuzz.trimf(HeightA.universe,[98.6,129.31,134])
HeightA['8'] = fuzz.trimf(HeightA.universe,[98.6,133.17,138])
HeightA['10'] = fuzz.trimf(HeightA.universe,[98.6,137.03,142])
HeightA['12'] = fuzz.trimf(HeightA.universe,[98.6,143.79,149])
HeightA['14'] = fuzz.trimf(HeightA.universe,[98.6,149.58,155])

WeightA['5'] = fuzz.trimf(WeightA.universe,[17,19.3,20])
WeightA['6'] = fuzz.trimf(WeightA.universe,[17,21.23,22])
WeightA['6X'] = fuzz.trimf(WeightA.universe,[17,22.2,23])
WeightA['7'] = fuzz.trimf(WeightA.universe,[17,24.13,25])
WeightA['8'] = fuzz.trimf(WeightA.universe,[17,27.02,28])
WeightA['10'] = fuzz.trimf(WeightA.universe,[17,31.85,33])
WeightA['12'] = fuzz.trimf(WeightA.universe,[17,37.64,39])
WeightA['14'] = fuzz.trimf(WeightA.universe,[17,43.43,45])

WaistA['5'] = fuzz.trimf(WaistA.universe,[45.05,51.15,53])
WaistA['6'] = fuzz.trimf(WaistA.universe,[45.05,53.08,55])
WaistA['6X'] = fuzz.trimf(WaistA.universe,[45.05,55.01,56])
WaistA['7'] = fuzz.trimf(WaistA.universe,[45.05,55.01,57])
WaistA['8'] = fuzz.trimf(WaistA.universe,[45.05,55.97,58])
WaistA['10'] = fuzz.trimf(WaistA.universe,[45.05,58.87,61])
WaistA['12'] = fuzz.trimf(WaistA.universe,[45.05,61.76,64])
WaistA['14'] = fuzz.trimf(WaistA.universe,[45.05,63.69,66])

conclusionA['5'] = fuzz.trimf(conclusionA.universe,[4.5,5,5.5])
conclusionA['6'] = fuzz.trimf(conclusionA.universe,[5.5,6,6.5])
conclusionA['6X'] = fuzz.trimf(conclusionA.universe,[6,6.5,7])
conclusionA['7'] = fuzz.trimf(conclusionA.universe,[6.5,7,7.5])
conclusionA['8'] = fuzz.trimf(conclusionA.universe,[7.5,8,9])
conclusionA['10'] = fuzz.trimf(conclusionA.universe,[9,10,11])
conclusionA['12'] = fuzz.trimf(conclusionA.universe,[11,12,13])
conclusionA['14'] = fuzz.trimf(conclusionA.universe,[13,14,14])

Arule1 = crtl.Rule(HeightA['6']&WeightA['6']&WaistA['6'],conclusionA['6'])
Arule2 = crtl.Rule(HeightA['7']&WeightA['7']&WaistA['7'],conclusionA['7'])
Arule3 = crtl.Rule(HeightA['5']&WeightA['5']&WaistA['5'],conclusionA['5'])
Arule4 = crtl.Rule(HeightA['6X']&WeightA['6X']&WaistA['6X'],conclusionA['6X'])
Arule5 = crtl.Rule(HeightA['8']&WeightA['8']&WaistA['8'],conclusionA['8'])
Arule6 = crtl.Rule(HeightA['10']&WeightA['10']&WaistA['10'],conclusionA['10'])
Arule7 = crtl.Rule(HeightA['12']&WeightA['12']&WaistA['12'],conclusionA['12'])
Arule8 = crtl.Rule(HeightA['14']&WeightA['14']&WaistA['14'],conclusionA['14'])
Arule9 = crtl.Rule(HeightA['14']&WeightA['14']&WaistA['14'],conclusionA['14'])
conclusioA_ctrl = crtl.ControlSystem([Arule1,Arule2,Arule3,Arule4,Arule5,Arule6,Arule7,Arule8,Arule9])
conclusioA = crtl.ControlSystemSimulation(conclusioA_ctrl)
conclusioA.input['HeightA'] = 125
conclusioA.input['WeightA'] = 22.1
conclusioA.input['WaistA'] = 55

conclusioA.compute()
print(conclusioA.output['conclusionA'])
conclusionA.view(sim=conclusioA)


HeightB['4-5'] = fuzz.trimf(HeightB.universe,[93.5,106.15,110])
HeightB['5-6'] = fuzz.trimf(HeightB.universe,[93.5,111.94,116])
HeightB['6-7'] = fuzz.trimf(HeightB.universe,[93.5,117.73,122])
HeightB['7-8'] = fuzz.trimf(HeightB.universe,[93.5,123.52,128])
HeightB['8-9'] = fuzz.trimf(HeightB.universe,[93.5,129.31,134])
HeightB['9-10'] = fuzz.trimf(HeightB.universe,[93.5,135.1,140])
HeightB['10-11'] = fuzz.trimf(HeightB.universe,[93.5,140.89,146])
HeightB['11-12'] = fuzz.trimf(HeightB.universe,[93.5,146.68,152])
HeightB['12-13'] = fuzz.trimf(HeightB.universe,[93.5,152.47,158])
HeightB['13-14'] = fuzz.trimf(HeightB.universe,[93.5,158.26,164])

ChestB['4-5'] = fuzz.trimf(ChestB.universe,[49.3,55.97,58])
ChestB['5-6'] = fuzz.trimf(ChestB.universe,[49.3,58.38,60.5])
ChestB['6-7'] = fuzz.trimf(ChestB.universe,[49.3,60.8,63])
ChestB['7-8'] = fuzz.trimf(ChestB.universe,[49.3,62.73,65])
ChestB['8-9'] = fuzz.trimf(ChestB.universe,[49.3,66.1,68.5])
ChestB['9-10'] = fuzz.trimf(ChestB.universe,[49.3,69,71.5])
ChestB['10-11'] = fuzz.trimf(ChestB.universe,[49.3,71.89,74.5])
ChestB['11-12'] = fuzz.trimf(ChestB.universe,[49.3,75.27,78])
ChestB['12-13'] = fuzz.trimf(ChestB.universe,[49.3,77.2,80])
ChestB['13-14'] = fuzz.trimf(ChestB.universe,[49.3,79.61,82.5])

WaistB['4-5'] = fuzz.trimf(WaistB.universe,[46.75,53.08,55])
WaistB['5-6'] = fuzz.trimf(WaistB.universe,[46.75,54.04,56])
WaistB['6-7'] = fuzz.trimf(WaistB.universe,[46.75,55.01,57])
WaistB['7-8'] = fuzz.trimf(WaistB.universe,[46.75,55.97,58])
WaistB['8-9'] = fuzz.trimf(WaistB.universe,[46.75,57.9,60])
WaistB['9-10'] = fuzz.trimf(WaistB.universe,[46.75,59.59,61.75])
WaistB['10-11'] = fuzz.trimf(WaistB.universe,[46.75,61.76,64])
WaistB['11-12'] = fuzz.trimf(WaistB.universe,[46.75,63.21,65.5])
WaistB['12-13'] = fuzz.trimf(WaistB.universe,[46.75,64.17,66.5])
WaistB['13-14'] = fuzz.trimf(WaistB.universe,[46.75,64.9,67.25])

conclusionB['4-5'] = fuzz.trimf(conclusionB.universe,[4,4,5])
conclusionB['5-6'] = fuzz.trimf(conclusionB.universe,[4,5.5,6])
conclusionB['6-7'] = fuzz.trimf(conclusionB.universe,[5,6.5,7])
conclusionB['7-8'] = fuzz.trimf(conclusionB.universe,[6,7.5,8])
conclusionB['8-9'] = fuzz.trimf(conclusionB.universe,[7,8.5,9])
conclusionB['9-10'] = fuzz.trimf(conclusionB.universe,[8,9.5,10])
conclusionB['10-11'] = fuzz.trimf(conclusionB.universe,[9,10.5,11])
conclusionB['11-12'] = fuzz.trimf(conclusionB.universe,[10,11.5,12])
conclusionB['12-13'] = fuzz.trimf(conclusionB.universe,[11,12.5,13])
conclusionB['13-14'] = fuzz.trimf(conclusionB.universe,[12,13.5,14])

Brule1 = crtl.Rule(HeightB['4-5']&ChestB['4-5']&WaistB['4-5'],conclusionB['4-5'])
Brule2 = crtl.Rule(HeightB['5-6']&ChestB['5-6']&WaistB['5-6'],conclusionB['5-6'])
Brule3 = crtl.Rule(HeightB['6-7']&ChestB['6-7']&WaistB['6-7'],conclusionB['6-7'])
Brule4 = crtl.Rule(HeightB['7-8']&ChestB['7-8']&WaistB['7-8'],conclusionB['7-8'])
Brule5 = crtl.Rule(HeightB['8-9']&ChestB['8-9']&WaistB['8-9'],conclusionB['8-9'])
Brule6 = crtl.Rule(HeightB['9-10']&ChestB['9-10']&WaistB['9-10'],conclusionB['9-10'])
Brule7 = crtl.Rule(HeightB['10-11']&ChestB['10-11']&WaistB['10-11'],conclusionB['10-11'])
Brule8 = crtl.Rule(HeightB['11-12']&ChestB['11-12']&WaistB['11-12'],conclusionB['11-12'])
Brule9 = crtl.Rule(HeightB['12-13']&ChestB['12-13']&WaistB['12-13'],conclusionB['12-13'])
Brule10 = crtl.Rule(HeightB['13-14']&ChestB['13-14']&WaistB['13-14'],conclusionB['13-14'])


conclusioB_ctrl = crtl.ControlSystem([Brule1,Brule2,Brule3,Brule4,Brule5,Brule6,Brule7,Brule8,Brule9,Brule10])
conclusioB = crtl.ControlSystemSimulation(conclusioB_ctrl)
conclusioB.input['HeightB'] = 125
conclusioB.input['ChestB'] = 61
conclusioB.input['WaistB'] = 55

conclusioB.compute()
print(conclusioB.output['conclusionB'])
conclusionB.view(sim=conclusioB)
```
